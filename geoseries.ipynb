```python
import geopandas as gpd

# Read your GeoJSON file
countries = gpd.read_file("countries.geojson")

# Preview first rows
countries.head()
# --- Attributes ---
print("\n--- Attributes ---")
print("Area:\n", countries.geometry.area)
print("\nBounds:\n", countries.geometry.bounds)
print("\nTotal bounds:\n", countries.geometry.total_bounds)
print("\nGeometry type:\n", countries.geometry.geom_type)
print("\nIs valid:\n", countries.geometry.is_valid)

# --- Basic methods ---
print("\n--- Basic methods ---")
# Distance: example - distance from first shape to all others
print("Distance from first geometry:\n", countries.geometry.distance(countries.geometry.iloc[0]))
print("\nCentroids:\n", countries.geometry.centroid)
print("\nRepresentative points:\n", countries.geometry.representative_point())

# Change CRS example (this assumes your data is in EPSG:4326)
countries_meters = countries.to_crs(epsg=3857)
print("\nCRS changed to EPSG:3857:\n", countries_meters.crs)

# --- Relationship tests ---
print("\n--- Relationship tests ---")
# geom_equals_exact: compare first geometry to itself
print("Geom equals exact (self compare):", countries.geometry.iloc[0].equals_exact(countries.geometry.iloc[0], tolerance=0.001))

# contains: check if first geometry contains second
print("Contains test:", countries.geometry.iloc[0].contains(countries.geometry.iloc[1]))

# intersects: check intersection between first and second
print("Intersects test:", countries.geometry.iloc[0].intersects(countries.geometry.iloc[1]))

# --- Plot example ---
countries.plot()
```
